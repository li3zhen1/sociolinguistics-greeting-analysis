```python
import os
import re

# 4 digits with no other digits before or after
reg = r'(?<!\d)\d{4}(?!\d)'
def get_year(file_name: str) -> int|None:
    if re.search(reg, file_name) is not None:
        return int(re.search(reg, file_name).group(0))
    else:
        return None



def list_all_files_matched(path: str):
    for root, dirs, files in os.walk(path):
        for file in files:
            yof = get_year(file)
            if yof is not None:
                print(f'{yof} - {file}')
```

```python
list_all_files_matched('./captions/1960s/')
```

```python
import os
import  chardet

def convert2utf8(path: str):
    for root, dirs, files in os.walk(path):
        for file in files:
            with open(os.path.join(root, file), 'rb') as f:
                content = f.read()

                # if file is ecoded in utf8 with BOM, remove BOM
                # if file.endswith('.srt'):
                if content.startswith(b'\xef\xbb\xbf'):
                    print(f'remove BOM from {file}')
                    content = content[3:]
                    with open(os.path.join(root, file), 'wb') as f:
                        f.write(content)
                # if file is encoded in GB2312, convert to utf8
                chardet_result = chardet.detect(content)
                if chardet_result['encoding'] is None:
                    continue

                if not chardet_result['encoding'].startswith('utf'):
                    print(f'converting {file} from {chardet_result["encoding"]} to utf8')
                    try:
                        _content = content.decode(chardet_result['encoding']).encode('utf8')
                    
                        with open(os.path.join(root, file), 'wb') as f:
                            f.write(_content)
                            print(f'    convert {file} done')
                    except:
                        print(f'    convert {file} failed')
                        with open(os.path.join(root, file), 'wb') as f:
                            f.write(content)
                            print(f'    recover {file} done')
                
                
convert2utf8('./captions/1970s/')
```
